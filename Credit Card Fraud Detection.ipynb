**`=> the data set is highly unbalanced`**

0 = normal transaction

1 = fradulant transaction

```python
# seprate the data fraud and legit
legit= df[df.Class == 0]
fraud= df[df.Class == 1]
```

**`=> under sampling`**

```python
legit_sample= legit.sample(n= 492)
```

```python
# concate the legit_sample and fraud then create new data frame 
df1= pd.concat([legit_sample, fraud], axis= 0)
```

```python
df1
```

```python
df1.shape
```

```python
df1.describe()
```

**`=> separate the data X and Y`**

```python
X= df1.drop(["Class"], axis= 1)
Y= df1["Class"]
```

```python
print(X)
```

```python
print(Y)
```

**`=> data convert into train, test & split`**

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, stratify= Y, random_state= 42, test_size= 0.2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**`=> build the model and train the model`**

```python
lr= LogisticRegression()
lr.fit(X_train, Y_train)
```

**`=> evoluation the model`**

```python
# check the accuracy score on train data
train_pred= lr.predict(X_train)

train_acc= accuracy_score(train_pred, Y_train)
print(train_acc)
```

```python
# check the accuracy score on test data
test_pred= lr.predict(X_test)

test_acc= accuracy_score(test_pred, Y_test)
print(test_acc)
```
